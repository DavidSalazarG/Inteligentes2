```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([5,2,4,9,3,7,10])
y = np.array([2.6,1.5,2.2,3.5,2,2.7,3.8])

media_x = np.mean(x)
media_y = np.mean(y)
std = np.std(x)
var = np.var(x)
cov = np.cov(x,y)
print("media x {} media y {} std {} var {} cov".format(media_x,media_y,std,var,cov))
print(cov)
```

```python
b1,b0=np.polyfit(x,y,1)
plt.scatter(x,y,color='blue',label="Datos")
plt.plot(x,b1*x+b0,color='red',label="Regresion")
plt.legend()
```

```python
x = np.array([5,2,4,9,3,7,10]).reshape(-1,1)
y = np.array([2.6,1.5,2.2,3.5,2,2.7,3.8])
modelo = LinearRegression()
modelo.fit(x,y)
R2 = modelo.score(x,y)
print(R2)
```

```python
x_n = np.array([[0],[30]])
plt.scatter(x,y,color='blue',label='datos')
y_g = modelo.predict(x_n)
plt.plot(x_n,y_g,color='red',label='prediccion')
```
